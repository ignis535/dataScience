### Multiple boolean series

```python
cond1 = df["순이익률(%)"] < df["영업이익률(%)"]
cond2 = df["PER(배)"] < 1
```

* and: &, or: |

* cython을 사용하기 때문

```python
final_cond = cond1 & cond2
final_cond.head()
```

```python
df[final_cond].head()
```

```python
# boolean condition으로 데이터를 추출한 경우, column을 추출할 때 loc을 사용해야한다.
df.loc[final_cond, ["ROE(%)"]].head()
```

### isin()

```python
name_list = ["삼성전자", "현대건설", "삼성물산"]
```

1. mutiple boolean series를 이용하는 방법

```python
cond1 = df["종목명"] == "삼성전자"
cond2 = df["종목명"] == "현대건설"
cond3 = df["종목명"] == "삼성물산"

final_cond = cond1 | cond2 | cond3
df.loc[final_cond]
```

2. index화 해서 loc으로 row-wise indexing을 이용해서 가져오는 방법

```python
tmp_df = df.set_index("종목명")
tmp_df.loc[name_list]
```

3. isin() 함수를 이용해서 가져오는 방법

```python
df.loc[df["종목명"].isin(name_list), ["종목명", "ROA(%)", "ROE(%)"]]
```

### all() vs any()

* boolean series의 각 row에 대하여

* 모두 true이면 all()은 true를 리턴하고, 그렇지 않으면 false

* 적어도 하나가 true이면 any()는 true를 리턴하고, 그렇지 않으면 false

```python
cond = df["순이익률(%)"] > 0
```

```python
cond.all()
```

```python
cond.any()
```

```python
(df["순이익률(%)"] > 0).all()
(df["순이익률(%)"] > 0).any()
```

## 연산

```python
import FinanceDataReader as fdr
price_df = fdr.DataReader("005930", "2009-09-16", "2018-03-21")
price_df.head()
```

### 연산 기준

* DataFrame은 기준이 column

* Series는 기준이 index

* 따로 명시가 없다면 Series의 index가 DataFrame의 column에 맞춰진다!

#### DataFrame & Series

```python
price_df.iloc[0]
```

```python
# subtract row series
# DataFrame의 기준인 column과 Sereis의 기준인 index가 서로 일치하기 때문에 의도한 연산 가능
(price_df - price_df.iloc[0]).head()
```

```python
# subtract column series
# DataFrame의 기준인 column와 Series의 기준인 index가 서로 일치하지 않기 때문에 의도한 연산 불가능
(price_df - price_df["Open"]).head()
```

* DataFrame에서 제공하는 연산관련 함수를 이용하면 의도한 연산을 수행할 수 있다.

* `sub()`의 경우 description에 "For Series input, axis to match Series index on"라고 써있다.

* axis = 0 or 1은 무조건 description(shift + tap)을 먼저 보고 판단하자

```python
close_series = price_df["Close"]
price_df.sub(close_series, axis = 0)
```

* "A simple brute froce solution of trying both directions until achieving the desired result is one possibility"

```python
price_df[["Open", "Close"]].sum(axis = 1).head()
price_df[["Open", "Close"]].sum(axis = 0).head()
```

#### DataFrame & DataFrame

* index, column이 일치하는 것만 element-wise 연산이 수행되고 나머지는 nan 처리

```python
price_df - price_df[["Open", "Low"]].iloc[ : 2]
```
